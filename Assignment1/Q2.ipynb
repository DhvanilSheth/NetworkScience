```python
import numpy as np
import matplotlib.pyplot as plt
from collections import deque
```

```python
edges = []
with open("C://Users//dhvan//Downloads//NetworkScience//Assignment1//directed_graph//p2p-Gnutella06.txt", "r") as f:
    for line in f:
        if line.startswith('#'):
            continue
        src, dst = map(int, line.strip().split())
        edges.append((src, dst))
```

 *In and Out degree of directed graphs.*


```python
# Determine the number of nodes
max_node = max(max(u, v) for u, v in edges)
N = max_node + 1

# Compute in-degree and out-degree
in_degree = [0]*N
out_degree = [0]*N
for (u, v) in edges:
    out_degree[u] += 1
    in_degree[v] += 1
```

*Plotting the In-Degree Distribution.*

```python
# Plot in-degree distribution (No. of Nodes vs In-degree)
unique_in_deg, counts = np.unique(in_degree, return_counts=True)
plt.figure(figsize=(8, 6))
plt.scatter(unique_in_deg, counts, s=10, color='orange', marker='o')
plt.title("In-Degree Distribution")
plt.xlabel("In-Degree")
plt.ylabel("No. of Nodes")
plt.savefig("in_degree_distribution.png")
plt.show()

# also plot it as a log-log plot
plt.figure(figsize=(8, 6))
plt.scatter(unique_in_deg, counts, s=10, color='green', marker='o')
plt.yscale('log')
plt.xscale('log')
plt.title("In-Degree Distribution (Log-Log)")
plt.xlabel("In-Degree")
plt.ylabel("No. of Nodes")
# plt.savefig("in_degree_distribution_loglog.png")
plt.show()
```

*Plotting the Out-Degree Distribution*


```python
# Plot out-degree distribution (No. of Nodes vs Out-degree)

unique_out_deg, counts = np.unique(out_degree, return_counts=True)
plt.figure(figsize=(8, 6))
plt.scatter(unique_out_deg, counts, s=10, color='blue', marker='o')
plt.title("Out-Degree Distribution")
plt.xlabel("Out-Degree")
plt.ylabel("No. of Nodes")
plt.savefig("out_degree_distribution.png")
plt.show()

# also plot it as a log-log plot
plt.figure(figsize=(8, 6))
plt.scatter(unique_out_deg, counts, s=10, color='red', marker='o')
plt.yscale('log')
plt.xscale('log')
plt.title("Out-Degree Distribution (Log-Log)")
plt.xlabel("Out-Degree")
plt.ylabel("No. of Nodes")
# plt.savefig("out_degree_distribution_loglog.png")
plt.show()
```
